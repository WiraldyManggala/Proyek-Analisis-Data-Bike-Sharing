## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
# Load datasets
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')
```

### Assessing Data

```python
print(day_df.info())
print(hour_df.info())
```

1. Dataset tidak memiliki missing values berdasarkan pengecekan awal.
2. Data terdiri dari berbagai fitur seperti tanggal, musim, suhu, kelembaban, dan jumlah peminjaman.
3. Tidak ada nilai yang perlu diisi ulang atau dihapus dalam tahap awal ini.

### Cleaning Data

```python
print(day_df.isnull().sum())
print(hour_df.isnull().sum())

print(day_df.duplicated().sum())
print(hour_df.duplicated().sum())
```

1. Tidak ditemukan data duplikat sehingga tidak ada data yang perlu dihapus.
2. Normalisasi nilai dapat dilakukan jika diperlukan untuk analisis lebih lanjut.


## Exploratory Data Analysis (EDA)

### Explore ...

```python
plt.figure(figsize=(10, 5))
sns.histplot(day_df['cnt'], bins=30, kde=True)
plt.title('Distribusi Total Peminjaman Sepeda Harian')
plt.show()
```

1. Distribusi jumlah peminjaman sepeda menunjukkan pola yang mendekati distribusi normal,dengan sebagian besar peminjaman berkisar dalam rentang tertentu.
2. Ada kemungkinan adanya hari-hari tertentu dengan jumlah peminjaman yang jauh lebih tinggi atau lebih rendah dari biasanya,yang dapat dikaitkan dengan faktor eksternal seperti cuaca, hari libur, atau event khusus.

## Visualization & Explanatory Analysis

### Bagaimana pengaruh cuaca terhadap jumlah peminjaman sepeda?

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='weathersit', y='cnt', data=day_df)
plt.title('Pengaruh Cuaca terhadap Jumlah Peminjaman')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Peminjaman')
plt.show()
```

Boxplot digunakan karena dapat menunjukkan distribusi data serta outlier yang mungkin ada.
1. Dari visualisasi ini terlihat bahwa pada kondisi cuaca lebih buruk, jumlah peminjaman sepeda cenderung lebih rendah dibandingkan dengan cuaca cerah.
2. Ini mengindikasikan bahwa kondisi cuaca secara signifikan memengaruhi jumlah peminjaman sepeda.

### Bagaimana tren peminjaman sepeda berdasarkan waktu (harian dan bulanan)?

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='mnth', y='cnt', data=day_df, estimator='mean')
plt.title('Tren Peminjaman Sepeda berdasarkan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Peminjaman')
plt.show()
```

Lineplot digunakan karena efektif untuk menunjukkan tren data dari waktu ke waktu.
1. Dari visualisasi ini, terlihat adanya tren musiman di mana jumlah peminjaman cenderung meningkat pada bulan-bulan tertentu, kemungkinan terkait dengan perubahan cuaca dan liburan.
2. Bulan-bulan musim panas tampaknya memiliki peminjaman yang lebih tinggi dibandingkan dengan musim dingin.

Peminjaman sepeda menunjukkan pola musiman, dengan peningkatan pada bulan-bulan tertentu.
